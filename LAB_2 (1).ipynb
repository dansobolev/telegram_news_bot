```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://feeds.feedburner.com/yandex/MAOo'
```

```python
resp = requests.get(url)
```

```python
soup = BeautifulSoup(resp.content, features = 'xml')
```

```python
items = soup.findAll('item')
```

```python
items[1]
```

```python
key_word = 'Россия'
for i in range(len(items)):    
    if (items[i].description.text).find(key_word) != -1:
        print(items[i].description.text)
```
